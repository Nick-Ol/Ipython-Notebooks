```python
import pandas as pd
import pylab as pylab
import mpld3
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
mpld3.enable_notebook()
```

```python
def get_signal(path_to_csv_file):
    sig = pd.read_csv(path_to_csv_file, squeeze=True).tolist()
    return sig

signal = get_signal("subject13_FPz_A1.csv")
```

```python
start = 1000000

x = signal[start:(start + 600*200)] 

NFFT = 200*10 #10 s
Fs = 200 #200 Hz

fig = plt.figure(figsize=(14, 10))
Pxx, freqs, bins, im = pylab.specgram(x, Fs = 200, NFFT = NFFT)
```

```python

    
path_to_write = "features.csv"
test = pd.DataFrame(np.matrix.transpose(Pxx), columns = freqs, index = bins)
test
```
