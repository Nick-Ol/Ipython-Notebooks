#Spectrograms of 1D signal in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline
```

##Arguments and outputs of specgram()

Reminder: frequency-pulsation conversion :

$\omega = 2 \pi f$

$f = \frac{\omega}{2 \pi}$

Two remarks: spectrogram is invariant if the signal is multiplied by a constant, and if a constant is added to the signal.

```python
dt = 0.0005 #1/dt = 2,000 Hz
t = arange(0, 20, dt)
freq1 = 100
freq2 = 200
freq3 = 500

mask1 = where(logical_and(t > 0.5, t < 3), 1.0, 0.0)
mask2 = where(logical_and(t > 2, t < 5), 1.0, 0.0)
mask3 = where(logical_and(t > 7.5, t < 13), 1.0, 0.0)
```

```python
s1 = mask1*sin(np.dot(2*pi*freq1, t))

s1 = np.add(s1, np.dot(3, mask2*sin(np.dot(2*pi*freq2, t))))
s1 = np.add(s1, np.dot(10, mask3*sin(np.dot(2*pi*freq3, t))))

noiseAmpl = 0.1
noise = noiseAmpl*rand(len(time))
s1 = np.add(noise, s1)

plot(t, s1)
```

specgram() divides the signal vector into overlapping segments, applies a window to each of them and forms a column with its discrete  Fourier transform.

Useful parameters of specgram():
- NFFT is the number of points used to compute the discrete Fourier transform (= the length of each segment). By default, it is min(256, len(signal)). A power 2 is said to be more efficient.
- Fs is the sampling frequency. It is used to scale the plot (max value is Fs/2, cf Nyquist theorem). By default, it is 2.
- window is the window applied to each segment. By default, it is a Hanning window (of length NFFT).
- noverlap is the number of samples each segment of the signal overlaps. By default, it is len(window)/2
- scale_by_frequency: boolean. If True, the y-axis will be scaled by Fs.

```python
NFFT = 256
spec1_a = specgram(s1, Fs = 1/dt, NFFT = NFFT, scale_by_freq = False, noverlap = NFFT/2)
```

We see our three sinusoids, with increasing frequencies (100, 200, 500 Hz), and increasing amplitudes (the stripes become redder and redder).

Even though the two first sin overlap, we see them clearly separated on the spectrogram.

We also see 6 side effects (at 0.5, 2, 3, 5, 7.5 and 13, which are the limits of our masks), each time a sin abruptly disappears.
The larger NFFT, the more important the side effect: indeed, see what happens with 1024 points/segment.

Explanation: the larger NFFT, the lesser segments there are, the wider the vertical stripes on the spectrogram.

```python
spec1_b = specgram(s1, Fs = 1/dt, NFFT = 1024, scale_by_freq = False)
```

Return values of specgram():
- $Pxx$ is the value of the Fourier coefs. it's a $len(freqs),len(bins)$ matrix, and corresponds to what is plotted.

- $freqs$ is the array of Fourier coefs computed : from $0$ to $\frac{Fs}{2}$, and there are $\frac{NFFT}{2} +1$.

- $bins$ is an array containing the centers of the segments. We have $len(bins) = E[\frac{len(signal)}{NFFT-noverlap}]$. The signal is truncated, to have a signal whose length a multiple of $NFFT$.

- $im$ is the image. 

```python
Pxx, freqs, bins, im = spec1_a
Pxx.shape
```

##Another signal, sum of two masked sin and important noise.

```python
s2_a = sin(2*pi*5*t)
s2_b = 2*sin(2*pi*20*t)

mask = where(logical_and(t>10, t<13), 1.0, 0.0)
s2_b = s2_b * mask

nse_ampl = 2
nse = nse_ampl*randn(len(t))
s2 = s2_a + s2_b + nse
```

Let's plot the noised signal in blue and the orginal one in red:

```python
plot(t, s2, color = 'blue')
plot(t, s2_a + s2_b, color = 'red')
```

```python
spec2 = specgram(s2, Fs = 10, NFFT = 512)#, scale_by_freq = True)
```

Interesting remark : if the noise is normal and its stddev is not too great, we still manage to see the original frequencies.

##Single sinus with varying amplitude

Now we would like to study a signal whose frequency is constant (100 Hz), and whose amplitude is quadratic in time :
$$s(t) = t^2 \, \mathrm{sin} (100*2\pi t) + 0.01\,\epsilon(t)$$

```python
s3 = t**2*sin(2*pi*100*t) + 0.01*randn(len(t))
plot(t, s3)
```

```python
spec4 = specgram(s3, Fs = 1/dt, NFFT = 256)
```

There's only one frequency as expected, but the 100 Hz stripe is redder as time increases, because the amplitude of the signal is increasing.

##Constant amplitude, increasing frequency
Now we want a signal whose amplitude is constant, but whose frequency increases with time.

For that we use $s(t) = \mathrm{sin} [(10 + 10 t) 2\pi t]$

So at $t = 20 s$ we expect to see a local frequency of 210. And we get ~400. Why.

```python
s5 = sin((2*pi*10*t + 10*2*pi*t**2))
plot(t[:1000], s5[:1000])
```

```python
spec5 = specgram(s5, Fs = 1/dt, NFFT = 256, scale_by_freq = False)
```
