# How to find the optimal number of clusters ?

## Some possible methods :

- "Rule of thumb" number found on Wikipedia : $k \approx \sqrt{n/2}$


- Elbow method : We compute the algorithm (say K-Means) for different Ks, and plot a measure for each result. We select K where the marginal difference will drop after it (inflexion point). It is not always identifiable. 
As measure, we can think about distortion.


- Maximise the separation


- Maximize the silhouette

## Let's try it ! Case of O-U time series

We will first take as example Ornstein-Uhlenbeck time series, and use the Euclidean distance. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#1-dimension Ornstein Ulhenbeck process :
def simuOrnsteinOne(l, lambd, mu, sigma, dt, X0) :
    #length l, dX = lambd * (mu - X)dt + sigma * dB * sqrt(dt)
    X = np.zeros(l)
    X[0] = X0
    for t in range(1,l):
        X[t] = X[t-1] + lambd*(X[t-1]-mu)*dt + sigma*np.random.randn(1)*np.sqrt(dt)
    return np.asarray(X)

#simulate n One-dimensional Ornstein processes of size l:
def simulationsOne(n, l):
    simus = []
    params = []
    for i in range(0,n):
        lambd = -np.random.uniform(0,1)
        mu = np.random.uniform(-1,1)
        sigma = np.random.uniform(-1,1)
        X0 = np.random.uniform(-1,1)
        X = simuOrnsteinOne(l, lambd, mu, sigma, 1, X0)
        simus.append(X)
        params.append([lambd, mu, sigma, X0])
    return np.asarray(simus), np.asarray(params)

def plotSimulationsOne(L):
    fig= plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    for i in range(0,len(L)):      
        ax.plot(L[i])
        
data = simulationsOne(100, 500)[0]
plotSimulationsOne(data)
```

We have 100 samples. The "rule of thumb" formula would give $K \approx 7$

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

distortionScores = []
separationScores = []
separationBisScores = []
silhouetteScores = []
N = len(data)

def separation(kmeans_centroids, kmeans_labels):
    clustersSizes = np.bincount(kmeans_labels)
    
    S= 0
    sumWeights = 0
    for i in range(0,len(kmeans_centroids)):
        for j in range(0,i):
            S = S + clustersSizes[i]*clustersSizes[j] \
            *np.linalg.norm(kmeans_centroids[i]-kmeans_centroids[j])
            sumWeights = sumWeights + clustersSizes[i]*clustersSizes[j]
    
    return S/sumWeights ;

def separationBis(kmeans_centroids, kmeans_labels):
    clustersSizes = np.bincount(kmeans_labels)
    
    S= 0
    mu = kmeans_centroids.mean(axis=0)
    for i in range(0,len(kmeans_centroids)):
        for j in range(0,i):
            S = S + clustersSizes[i]*np.linalg.norm(kmeans_centroids[i]-mu)
    
    return S/np.sum(clustersSizes);

#let's plot :
for k in range(2,21): #2 to 20 clusters
    estimator = KMeans(init='k-means++', n_clusters=k, n_init=10)
    estimator.fit(data) 
    distortionScore = estimator.inertia_ / N
    separationScore = separation(estimator.cluster_centers_, estimator.labels_)
    separationBisScore = separationBis(estimator.cluster_centers_, estimator.labels_)
    silhouetteScore = metrics.silhouette_score(data, estimator.labels_,metric='euclidean',sample_size=N)
    
    distortionScores.append(distortionScore)
    separationScores.append(separationScore)
    separationBisScores.append(separationBisScore)
    silhouetteScores.append(silhouetteScore)
```

### Elbow method using distortion :

```python
plt.plot(range(2,21),distortionScores)
```

```python
#Marginal gains :
def marginalGains(A):
    Gains = []
    n = len(A)
    for i in range(1,n):
        Gains.append((A[i]-A[i-1])/A[i-1])
        
    return Gains
```

```python
plt.plot(range(2,20),marginalGains(distortionScores))
```

The marginal gain becomes low (in absolute value) after the passage from 7 to 8 clusters. We would then select K = 8

### Separation :

```python
plt.plot(range(2,21),separationScores)
```

```python
plt.plot(range(2,21),separationBisScores)
```

We have a maximum for K = 19 in one case, K = 20 in another... Not really meaningful here.

### Silhouette :

```python
plt.plot(range(2,21),silhouetteScores)
```

```python
np.argmax(silhouetteScores)
```

The silhouette is maximal for K = 3

## In $\mathbb{R}^2$ :

```python
dataTwo1 = np.random.multivariate_normal([1,2], np.eye(2), 100).tolist()
dataTwo2 = np.random.multivariate_normal([8,1], np.matrix([[2, 1],[2, 4]]), 100).tolist()
dataTwo3 = np.random.multivariate_normal([-3,0], np.eye(2), 100).tolist()
dataTwo4 = np.random.multivariate_normal([1,10], np.matrix([[8, 1],[2, 1]]),100).tolist()
dataTwo = np.asarray(dataTwo1 + dataTwo2 + dataTwo3 + dataTwo4)
```

```python
plt.scatter(dataTwo[:,0],dataTwo[:,1])
```

We would expect K = 3 or 4

```python
def plotKMeansRTwo(data, kmeans_labels, kmeans_cluster_centers):
    n = len(data)   
    K = len(kmeans_cluster_centers)
    colors = []
    #generate the colors vector :
    for i in range(K):
        colors.append('#%06X' % np.random.randint(0, 0xFFFFFF))
    fig= plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    for i in range(0, n):
        cluster = kmeans_labels[i]
        ax.scatter(data[i,0], data[i,1], color=colors[cluster])
    for j in range(0, K):
        ax.scatter(kmeans_cluster_centers[j,0], kmeans_cluster_centers[j,1], color='k', linewidth=2.0)
        #clusters curves in bold
```

```python
estimator = KMeans(init='k-means++', n_clusters=3, n_init=10)
estimator.fit(dataTwo)
plotKMeansRTwo(dataTwo, estimator.labels_, estimator.cluster_centers_)
```

```python
estimator = KMeans(init='k-means++', n_clusters=4, n_init=10)
estimator.fit(dataTwo)
plotKMeansRTwo(dataTwo, estimator.labels_, estimator.cluster_centers_)
```

```python
distortionScoresTwo = []
separationScoresTwo = []
separationBisScoresTwo = []
silhouetteScoresTwo = []
N = len(dataTwo)

#let's plot :
for k in range(2,21): #2 to 20 clusters
    estimator = KMeans(init='k-means++', n_clusters=k, n_init=10)
    estimator.fit(dataTwo) 
    distortionScoreTwo = estimator.inertia_ / N
    separationScoreTwo = separation(estimator.cluster_centers_, estimator.labels_)
    separationBisScoreTwo = separationBis(estimator.cluster_centers_, estimator.labels_)
    silhouetteScoreTwo = metrics.silhouette_score(dataTwo, estimator.labels_,metric='euclidean',sample_size=N)
    
    distortionScoresTwo.append(distortionScoreTwo)
    separationScoresTwo.append(separationScoreTwo)
    separationBisScoresTwo.append(separationBisScoreTwo)
    silhouetteScoresTwo.append(silhouetteScoreTwo)
```

### Elbow method using distortion :

```python
plt.plot(range(2,21),distortionScoresTwo)
```

```python
plt.plot(range(2,20),marginalGains(distortionScoresTwo))
```

The marginal gain is reduced after the passage from 4 to 5 clusters. It would give K = 5

```python
estimator = KMeans(init='k-means++', n_clusters=5, n_init=10)
estimator.fit(dataTwo)
plotKMeansRTwo(dataTwo, estimator.labels_, estimator.cluster_centers_)
```

### Separation :

```python
plt.plot(range(2,21),separationScoresTwo)
```

Clearly here, K = 3

```python
plt.plot(range(2,21),separationBisScoresTwo)
```

```python
plt.plot(range(2,20),marginalGains(separationBisScoresTwo))
```

The marginal gain is important forthe passage for 2 to 3, but then it decreases a lot. We would also have K = 3

### Silhouette :

```python
plt.plot(range(2,21),silhouetteScoresTwo)
```

Clearly, K = 3
