#K-Medoids on time series using Dynamic Time Warping

To compare time series of the same size, we could use Euclidean distance. It would calculate the distance between two points at the same time, for the wholes signals. The problem with this calculus is that two signals which are only dephased may have a large distance, whereas we want them to be similar : DTW address this issue. Furthermore, DTW allows to compare two signales of different size.

##Dynamic Time Warping

The dynamic time warping definite a matrix which is filled with the Euclidean distance between each points of the signals. Then, we will select the path which give the minimum of these distances' sum from the beginnings of the signales to their ends.

```python
import numpy as np

def DTWdistance(s1, s2):
    DTW={}
 
    for i in range(len(s1)):
        DTW[(i, -1)] = float('inf') #we fill the exterior of the matrix with infinite values
    for i in range(len(s2)):
        DTW[(-1, i)] = float('inf')
    DTW[(-1, -1)] = 0
 
    for i in range(len(s1)):
        for j in range(len(s2)):
            dist= np.abs(s1[i]-s2[j]) #we use the L1 norm
            DTW[(i, j)] = dist + min(DTW[(i-1, j)],DTW[(i, j-1)], DTW[(i-1, j-1)])
            #the distance at this stage is the distance between points i and j
            #plus the sum of distances over optimal path to get their
 
    return DTW[len(s1)-1, len(s2)-1] #the final distance is in the final case, which compare both ends of the signals
```

Let's try it on Ornstein Uhlenbeck processes.

```python
import matplotlib.pyplot as plt
%matplotlib inline

#1-dimension Ornstein Ulhenbeck process :
def simuOrnsteinOne(l, lambd, mu, sigma, dt, X0) :
    #length l, dX = lambd * (mu - X)dt + sigma dW
    X = np.zeros(l)
    X[0] = X0
    for t in range(1,l):
        X[t] = X[t-1] + lambd*(X[t-1]-mu)*dt + sigma*np.random.randn(1)*np.sqrt(dt)
    return np.asarray(X)

#simulate n One-dimensional Ornstein processes of size l:
def simulationsOne(n, l):
    simus = []
    params = []
    for i in range(0,n):
        lambd = -np.random.uniform(0,1)
        mu = np.random.uniform(-1,1)
        sigma = np.random.uniform(-1,1)
        X0 = np.random.uniform(-1,1)
        X = simuOrnsteinOne(l, lambd, mu, sigma, 1, X0)
        simus.append(X)
        params.append([lambd, mu, sigma, X0])
    return np.asarray(simus), np.asarray(params)

def plotSimulationsOne(L):
    fig= plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    for i in range(0,len(L)):      
        ax.plot(L[i])
        
S = simulationsOne(2,500)
plotSimulationsOne(S[0])
```

```python
print("L1 distance = %f" %np.linalg.norm(S[0][1]-S[0][0],1))
print ("DTW distance =  %f" %DTWdistance(S[0][0],S[0][1]))
```

##K-Medoids on time series

The problem woth K-Means with DTW distance, is the averaging over a cluster to compute the centroid. If we simply take the average, two depahsed signals, closed from eachother with DTW, will be far from their centroid...

The idea of K-Metroids is to selection as centroid a data point from the cluster, whose average dissimilarity to all the ohter points of the same cluster is minimal i.e. the most centrally located point in the cluster. 

```python
import numpy as np
import random

def KMedoids(data, K, maxIter, epsilon=.01):
    #K : number of clusters
    #maxIter : maximum iterations
    #epsilon : stopping criterion
    
    N = len(data)
    medoids_indices = random.sample(range(N), K) #initial medoids indices
    clusters = np.zeros(N) #will contain in which cluster is i-th point
    distances_to_medoids = np.zeros(K) #for each point, distances to the medoids
    cluster_distortion = np.zeros(K) #distortions intra-cluster
    distortions = [0] #distortions after each iteration
    iter = 0
    
    while True:
        for i in range(N):
            for j in range(K):
                distances_to_medoids[j] = DTWdistance(data[i], data[medoids_indices[j]])
            
            clusters[i] = int(np.argmin(distances_to_medoids)) #assign each point to its closest medoid
    
        for k in range(K):       
            indices_cluster_k = [i for i in range(N) if clusters[i]==k]
            intra_cluster_dist = np.zeros(len(indices_cluster_k))
            for i in range(len(indices_cluster_k)):
                for j in range(len(indices_cluster_k)):
                    if (i != j):
                        intra_cluster_dist[i] = intra_cluster_dist[i] \
                        + DTWdistance(data[indices_cluster_k[i]], data[indices_cluster_k[j]])
                        
            min_index = np.argmin(intra_cluster_dist) #most central point's index
            medoids_indices[k] = indices_cluster_k[min_index] 
            cluster_distortion[k] = np.min(intra_cluster_dist)
            
        distortions.append(np.sum(cluster_distortion))
        
        if ((abs(distortions[iter+1]-distortions[iter]) < epsilon) or (iter >= maxIter -1)):
            break
        
        iter = iter + 1;
        
    return medoids_indices, clusters, distortions, cluster_distortion
```

###Speed improvement

###1st step : do not recalculate every distances

```python
def non_repeat_DTW1(matrix, data, i, j):
    if (matrix[i][j] == -1): #if it has already been calculated
        matrix[i][j] = DTWdistance(data[i], data[j])
    return matrix[i][j]
    
def KMedoids_non_repeat(data, K, maxIter, epsilon=.01):
    #K : number of clusters
    #maxIter : maximum iterations
    #epsilon : stopping criterion
    
    N = len(data)
    medoids_indices = random.sample(range(N), K) #initial medoids indices
    clusters = np.zeros(N) #will contain in which cluster is i-th point
    distances_to_medoids = np.zeros(K) #for each point, distances to the medoids
    cluster_distortion = np.zeros(K) #distortions intra-cluster
    distortions = [0] #distortions after each iteration
    distMatrix = np.eye(N,N) - 1 #distances matrix
    iter = 0
    
    while True:
        for i in range(N):
            for j in range(K):
                distances_to_medoids[j] = non_repeat_DTW1(distMatrix, data, i, medoids_indices[j])
            
            clusters[i] = int(np.argmin(distances_to_medoids)) #assign each point to its closest medoid
    
        for k in range(K):       
            indices_cluster_k = [i for i in range(N) if clusters[i]==k]
            intra_cluster_dist = np.zeros(len(indices_cluster_k))
            for i in range(len(indices_cluster_k)):
                for j in range(len(indices_cluster_k)):
                    if (i != j):
                        intra_cluster_dist[i] = intra_cluster_dist[i] \
                        + non_repeat_DTW1(distMatrix, data, indices_cluster_k[i], indices_cluster_k[j])
                        
            min_index = np.argmin(intra_cluster_dist) #most central point's index
            medoids_indices[k] = indices_cluster_k[min_index] 
            cluster_distortion[k] = np.min(intra_cluster_dist)
            
        distortions.append(np.sum(cluster_distortion))
        
        if ((abs(distortions[iter+1]-distortions[iter]) < epsilon) or (iter >= maxIter -1)):
            break
        
        iter = iter + 1;
        
    return medoids_indices, clusters, distortions, cluster_distortion
```

```python
import time
t0 = time.clock()
kmed2 = KMedoids_non_repeat(data, 3, 100)
print time.clock() - t0, "seconds process time"
print kmed2[1]
print kmed2[2]
```

###2nd step : range the paths for the DTW computation

```python
def DTW_distance_ranged1(s1, s2, l1, l2):
    # l1 : range according to s1
    # l2 : range according to s2
    n1 = len(s1)
    n2 = len(s2)
    DTW = {}
    for i in range(n1):
        DTW[(i, -1)] = float('inf')
    for j in range(n2):
        DTW[(-1, j)] = float('inf')
    DTW[(-1, -1)] = 0
    for i in range(0, n2-l2):
        for j in range(l2+i, n2):
            DTW[(i, j)] = float('inf')  # upper right triangle
    for i in range(l1, n1):
        for j in range(0, i-l1+1):
            DTW[(i, j)] = float('inf')  # lower left triangle

    for i in range(0, len(s1)):
        for j in range(max(0, i-l1+1), min(len(s2), i+l2)):
            dist = np.abs(s1[i]-s2[j])
            DTW[(i, j)] = dist + \
                min(DTW[(i-1, j)], DTW[(i, j-1)], DTW[(i-1, j-1)])

    return DTW[len(s1)-1, len(s2)-1]

def non_repeat_DTW2(matrix, data, i, j, l1, l2):
    if (matrix[i][j] == -1):
        matrix[i][j] = DTW_distance_ranged1(data[i], data[j], l1, l2)
    return matrix[i][j]

def KMedoids_ranged(data, K, semi_width, maxIter, epsilon=1):

    N = len(data)
    medoids_indices = random.sample(range(N), K)
    clusters = np.zeros(N)
    distances_to_medoids = np.zeros(K)
    cluster_distortion = np.zeros(K)
    distortions = [0]
    distMatrix = np.eye(N, N) - 1
    iter = 1

    while True:
        for i in range(N):
            for j in range(K):
                distances_to_medoids[j] = \
                    non_repeat_DTW2(distMatrix, data, i,
                                   medoids_indices[j],
                                   semi_width, semi_width)

            clusters[i] = int(np.argmin(distances_to_medoids))

        for k in range(K):
            indices_cluster_k = [i for i in range(N) if clusters[i] == k]
            intra_cluster_dist = np.zeros(len(indices_cluster_k))
            for i in range(len(indices_cluster_k)):
                for j in range(len(indices_cluster_k)):
                    if (i != j):
                        intra_cluster_dist[i] = intra_cluster_dist[i] + \
                            non_repeat_DTW2(distMatrix, data,
                                           indices_cluster_k[i],
                                           indices_cluster_k[j],
                                           semi_width, semi_width)

            min_index = np.argmin(intra_cluster_dist)
            medoids_indices[k] = indices_cluster_k[min_index]
            cluster_distortion[k] = np.min(intra_cluster_dist)

        if (iter == 1):
            distortions = [np.sum(cluster_distortion)]
            if (iter >= maxIter-1):
                break
        else:
            distortions.append(np.sum(cluster_distortion))
            if ((abs(distortions[iter-1]-distortions[iter-2]) < epsilon) or
               (iter >= maxIter-1)):
                break

        iter = iter + 1

    return medoids_indices, clusters, distortions, cluster_distortion
```

```python
import time
t0 = time.clock()
kmed3 = KMedoids_ranged(data, 3, 25, 100)
print time.clock() - t0, "seconds process time"
print kmed3[1]
print kmed3[2]
```

###3rd step : add a pattern for the distances matrix

```python
def pattern_DTW(n1, n2, l1, l2):
    DTW = {}

    for i in range(n1):
        DTW[(i, -1)] = float('inf')
    for j in range(n2):
        DTW[(-1, j)] = float('inf')
    DTW[(-1, -1)] = 0
    for i in range(0, n2-l2):
        for j in range(l2+i, n2):
            DTW[(i, j)] = float('inf')  # upper right triangle
    for i in range(l1, n1):
        for j in range(0, i-l1+1):
            DTW[(i, j)] = float('inf')  # lower left triangle

    return DTW


def DTW_distance_ranged(pattern, s1, s2, l1, l2):
    # l1 : range according to s1
    # l2 : range according to s2
    DTW = pattern

    for i in range(0, len(s1)):
        for j in range(max(0, i-l1+1), min(len(s2), i+l2)):
            dist = np.abs(s1[i]-s2[j])
            DTW[(i, j)] = dist + \
                min(DTW[(i-1, j)], DTW[(i, j-1)], DTW[(i-1, j-1)])

    return DTW[len(s1)-1, len(s2)-1]


def non_repeat_DTW(pattern, matrix, data, i, j, l1, l2):
    if (matrix[i][j] == -1):
        matrix[i][j] = DTW_distance_ranged(pattern, data[i], data[j], l1, l2)
    return matrix[i][j]

def KMedoids_ranged2(data, K, semi_width, maxIter, epsilon=1):

    N = len(data)
    medoids_indices = random.sample(range(N), K)
    clusters = np.zeros(N)
    distances_to_medoids = np.zeros(K)
    cluster_distortion = np.zeros(K)
    distortions = [0]
    distMatrix = np.eye(N, N) - 1
    # we suppose the data all have the same dimension
    pattern = pattern_DTW(len(data[0]), len(data[0]), semi_width, semi_width)
    iters = 1

    while True:
        for i in range(N):
            for j in range(K):
                distances_to_medoids[j] = \
                    non_repeat_DTW(pattern, distMatrix, data, i,
                                   medoids_indices[j],
                                   semi_width, semi_width)

            clusters[i] = int(np.argmin(distances_to_medoids))

        for k in range(K):
            indices_cluster_k = [i for i in range(N) if clusters[i] == k]
            intra_cluster_dist = np.zeros(len(indices_cluster_k))
            for i in range(len(indices_cluster_k)):
                for j in range(len(indices_cluster_k)):
                    if (i != j):
                        intra_cluster_dist[i] = intra_cluster_dist[i] + \
                            non_repeat_DTW(pattern, distMatrix, data,
                                           indices_cluster_k[i],
                                           indices_cluster_k[j],
                                           semi_width, semi_width)

            min_index = np.argmin(intra_cluster_dist)
            medoids_indices[k] = indices_cluster_k[min_index]
            cluster_distortion[k] = np.min(intra_cluster_dist)

        if (iters == 1):
            distortions = [np.sum(cluster_distortion)]
            if (iters >= maxIter-1):
                break
        else:
            distortions.append(np.sum(cluster_distortion))
            if ((abs(distortions[iters-1]-distortions[iters-2]) < epsilon) or
               (iters >= maxIter-1)):
                break

        iters = iters + 1

    return medoids_indices, clusters, distortions, cluster_distortion
```

```python
import time
t0 = time.clock()
kmed4 = KMedoids_ranged2(data, 3, 25, 100)
print time.clock() - t0, "seconds process time"
print kmed4[1]
print kmed4[2]
```
