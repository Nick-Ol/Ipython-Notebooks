# How to evaluate the performance of a clustering algorithm ?

We will only place ourselves in the case of Dreem, where the data are unsupervised.
But we could still imagine we want to compare the results of two partitions : then, other criterions (such as homogeneity, completeness, v-measure, adjusted rand score or adjusted mutual information could be used).

## Theoritical measures

Below are criterions to evaluate the performance of clustering :

- inertia : intra-clusters similarity. It tells us if data are closed or not to their cluster's centroid (i.e. if points within a cluster are similar)
\begin{equation*} Inertia = \sum_{i=1}^K \sum_{x \in C_i} ||x-\mu_i||^2 \end{equation*}
We also have the distortion as the average of inertia over the dataset
\begin{equation*} Distortion = Inertia / N_{samples} \end{equation*}

These criterions are decreasing with the number of clusters. But it is a good way to compare to clustering algorithm for an identical number of clusters.
The better the clustering, the lower inertia.

- separation : inter-clusters similarity. It tells us whether clusters are well disjoint (i.e. if points from different clusters are dissimilar). It is calculated as the weighted average distance between clusters
\begin{equation*} Separation = \sum_{i \ne j} N_{c_i} N_{c_j} ||\mu_i - \mu_j|| / \sum_{i \ne j} N_{c_i} N_{c_j} \end{equation*}
The better the clustering, the higher the separation.
We can also give this definition : 
\begin{equation*} SeparationBis = \sum_{i =1}^K N_{c_i} ||\mu_i - \mu|| / N \end{equation*} where $\mu$ is the global centroid

- silhouette : the silhouette is a good criterion to compare both intra and inter cluster similarity. For a given point i, we compute :

a(i) = mean distance between i and points in the same class

b(i) = mean distance between i and points of the nearest cluster (from i)
Then, the silhouette of i is : \begin{equation*} s(i) = \frac{b(i)-a(i)}{max(a(i), b(i))}  \end{equation*}

For a given sample, we have \begin{equation*} -1 \leq s \leq 1  \end{equation*}

When s = -1, it means the point is very close to another cluster. s=1 means the point is right in its cluster. To have the global silhouette measure, we take the average of s(i) over the dataset :
\begin{equation*} Silhouette = \sum_{i = 1}^N s_i /N  \end{equation*}

Remark : the Silhouette is also a good way to find the ideal number of clusters. If we plot the sihouette in function of the number of clusters used, we may find a local maxima.

- size distribution : we can also want the clusters to be homogeneously distributed

## Computation

We will study these criterions over a datset of one-dimensional Ornsteint-Uhlenbeck processes (which can be compared to EEGs). 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

#1-dimension Ornstein Ulhenbeck process :
def simuOrnsteinOne(l, lambd, mu, sigma, dt, X0) :
    #length l, dX = lambd * (mu - X)dt + sigma dW
    X = np.zeros(l)
    X[0] = X0
    for t in range(1,l):
        X[t] = X[t-1] + lambd*(X[t-1]-mu)*dt + sigma*np.random.randn(1)*np.sqrt(dt)
    return np.asarray(X)
    
#simulate n One-dimensional Ornstein processes of size l:
def simulationsOne(n, l):
    simus = []
    params = []
    for i in range(0,n):
        lambd = -np.random.uniform(0,1)
        mu = np.random.uniform(-1,1)
        sigma = np.random.uniform(-1,1)
        X0 = np.random.uniform(-1,1)
        X = simuOrnsteinOne(l, lambd, mu, sigma, 1, X0)
        simus.append(X)
        params.append([lambd, mu, sigma, X0])
    return np.asarray(simus), np.asarray(params)
```

```python
data = simulationsOne(100, 500)[0]
```

Let's compute the KMeans for K = 5 (with an Euclidian measure, it is an example. DTW may be prefered) :

```python
from sklearn.cluster import KMeans

K = 5
kmeans = KMeans(init='k-means++', n_clusters=K, n_init=10)
#kmeans++ = select initial centroids in a "smart" way
kmeans.fit(data)    

def plotKMeans(data, kmeans_labels, kmeans_cluster_centers):
    n = len(data)   
    K = len(kmeans_cluster_centers)
    colors = []
    #generate the colors vector :
    for i in range(K):
        colors.append('#%06X' % np.random.randint(0, 0xFFFFFF))
    fig= plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    for i in range(0, n):
        cluster = kmeans_labels[i]
        ax.plot(data[i], color=colors[cluster])
    for j in range(0, K):
        ax.plot(kmeans_cluster_centers[j], color=colors[j], linewidth=2.0)
        #clusters curves in bold
        
plotKMeans(data, kmeans.labels_, kmeans.cluster_centers_)
```

Now, let's evaluate the inertia/distortion, separation and silhouette :

```python
# Inertia :
kmeans.inertia_
```

```python
# Distortion :
kmeans.inertia_/len(data)
```

```python
#Separation :
def separation(kmeans_centroids, kmeans_labels):
    clustersSizes = np.bincount(kmeans_labels)
    
    S= 0
    sumWeights = 0
    for i in range(0,len(kmeans_centroids)):
        for j in range(0,i):
            S = S + clustersSizes[i]*clustersSizes[j] \
            *np.linalg.norm(kmeans_centroids[i]-kmeans_centroids[j])
            sumWeights = sumWeights + clustersSizes[i]*clustersSizes[j]
    
    return S/sumWeights ;

separation(kmeans.cluster_centers_, kmeans.labels_)
```

```python
#Separation, second formula :
def separationBis(kmeans_centroids, kmeans_labels):
    clustersSizes = np.bincount(kmeans_labels)
    
    S= 0
    mu = np.mean(kmeans_centroids)
    for i in range(0,len(kmeans_centroids)):
        for j in range(0,i):
            S = S + clustersSizes[i]*np.linalg.norm(kmeans_centroids[i]-mu)
    
    return S/np.sum(clustersSizes);

separationBis(kmeans.cluster_centers_, kmeans.labels_)
```

```python
#Silhouette :
from sklearn import metrics
silhouette = metrics.silhouette_score(data, kmeans.labels_,metric='euclidean',sample_size=len(data))
print ("Silhouette globale = %f" %silhouette) 
```

```python
#with less points taken into account (faster computation):
from __future__ import division
silhouette50 = metrics.silhouette_score(data, kmeans.labels_,metric='euclidean',sample_size=len(data)/2)
silhouette10 = metrics.silhouette_score(data, kmeans.labels_,metric='euclidean',sample_size=len(data)/2)
print ("Silhouette sur 50%% des données = %f \n Silhouette sur 10%% = %f" %(silhouette50, silhouette10)) 
```

## Silhouette for a different number of clusters

```python
silhouetteScores = []
for k in range(2,21):
    estimator = KMeans(init='k-means++', n_clusters=k, n_init=10)
    estimator.fit(data) 
    silhouetteScores.append(metrics.silhouette_score(data, estimator.labels_,metric='euclidean',sample_size=len(data)))

plt.plot(range(2,21),silhouetteScores)
```
