#An informal, illustrated explanation of PCA in 2D and 3D


We are going to illustrate how a Principal Component Analysis work, on simple data in 2D and 3D.

PCA is often used to reduce the dimensionality of real-world data (for example, before applying clustering algorithms).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

%matplotlib inline
```

```python
np.random.seed(1)


def generateEllipsis(mean, cov, cardinal, isNoised):
    data = np.random.multivariate_normal(mean, cov, cardinal)
    if (isNoised):
        data = data + np.random.randn(cardinal, len(mean))
    lambd, axes = np.linalg.eig(cov)
    return data, axes, lambd


def plotEllipsis(data, axes):
    plt.axis("equal")
    plt.scatter(data[:,0], data[:, 1], color='red', s=1)    
    for i in range(len(axes)):
        plt.quiver(axes[0, i], axes[1, i], scale = 2.5, color='blue')       

        
def plot2DComponents(pca):
    components = pca.components_
    for comp in components:
       plt.quiver(comp[0], comp[1], scale=5, color='green')
```

Now we're going to generate pseudo random data (2D Gaussian). It'll be centered. For covariance matrix, we take 
$$\pmatrix{2 & -1\\-1 & 2}$$
If you want to try another ellipsis, note that the covariance matrix must be symmetrical, but also definite positive.


```python
mean = [0, 0]
cov1 = np.matrix([[2, -1], [-1, 2]])

ellipsis1, axes1, lambd1 = generateEllipsis(mean, cov1, 1000, isNoised=False)
plotEllipsis(ellipsis1, axes1)

pca1 = PCA(n_components=2)
pca1.fit(ellipsis1)
plot2DComponents(pca1)

```

```python
print(pca1.explained_variance_ratio_) 
```

In the non-noised case, the first component is responsible for 75 % of the variance, and the second explains the 25 % remaining.

##General 2D case

We need to define a 
For this we define an orthonormal matrix, P. In dimension 2 it is very easy: $$P = \frac{1}{a^2 + b^2}\pmatrix{a & -b \\ b &a}$$ 

Then $P^{-1} = P^{T}$, and with $D = \pmatrix{\lambda_1 >0 & 0 \\ 0 & \lambda_2>0}$, we take as covariance matrix $PDP^{-1}$

```python
P = np.matrix([[1, -4], [4, 1]])
P = P/np.linalg.norm(P)
D = np.diag([1, 9])
cov2 = P*D*np.transpose(P)
print(cov2)
```

Wy do we get ellipsis ?
The general equation of the levels lines of our 2D Gaussian distribution is:
$$X^T\Sigma^{-1}X = c \iff X^T P^{-1}D^{-1}PX = c \iff (PX)^T \pmatrix{\frac{1}{\lambda_1} & 0 \\ 0 & \frac{1}{\lambda_2}} (PX) = c \iff \frac{y_1^2}{\lambda_1} + \frac{y_2^2}{\lambda_2} = c  $$ where $Y = \pmatrix{y_1 \\ y_2} = PX$.

As $P$ is orthonormal, $Y = PX$ is just the equation of a change of axis.

```python
ellipsis2, axes2, lambd2 = generateEllipsis(mean, cov2, 1000, isNoised=False)

plotEllipsis(ellipsis2, axes2)

pca2 = PCA(n_components=2)
pca2.fit(ellipsis2)
plot2DComponents(pca2)
```

```python
print(pca2.explained_variance_ratio_)
```

The variance explained by the i-th component is proportional to $\lambda_i$ (where the eigenvalues are in descending order: $\lambda_1 \geq ... \geq \lambda_n$). Feel free to change the values of $D$ :)

In general there are two approaches to PCA: 

- say "I want to keep $p$ components" (out of $n$ in the beginning, i.e. go from $\mathbb{R}^n$ to $\mathbb{R}^p$), and see how much of the variance is conserved 
- say "I want to keep $x$ % of the variance", without fixing the number of components.

It's not the best to see in 2D as there are at most 2 components ($p\leq n$), so let's study it in 3D:

##PCA in 3D

```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
from mpl_toolkits.mplot3d import Axes3D
```

```python
class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

```python
axes = [axis/np.linalg.norm(axis) for axis in [[1, 0, 1], [-1, 1, 1], [1, 2, -1]] ]
#normalized axes otherwise eigenvalues are modified

P_3D = np.matrix(axes)
P_3D = P_3D/np.linalg.norm(P_3D)
D_3D = np.diag([1, 3, 6])
cov_3D = P_3D*D_3D*np.transpose(P_3D)
```

```python
mean_3D = np.asarray([0, 0, 0])
ellipsis3, axes3, lambd3 = generateEllipsis(mean_3D, cov_3D, 5000, isNoised=False)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.axis("equal")
ax.set_xlabel('variable X')
ax.set_ylabel('variable Y')
ax.set_zlabel('variable Z')
ax.plot(ellipsis3[:,0], ellipsis3[:,1], ellipsis3[:,2], 'o', markersize=1, color='green', alpha=0.2)
    
scale = 2
pca3D = PCA(n_components = 3)
pca3D.fit(ellipsis3)
components = pca3D.components_
for comp in components:
    V = Arrow3D([mean_3D[0], scale*comp[0]], [mean_3D[1], scale*comp[1]], [mean_3D[2], scale*comp[2]], lw=3, arrowstyle="-|>", color="blue")
    ax.add_artist(V)
plt.show()
```

```python
print(pca3D.explained_variance_ratio_)
print(components)
```

2 options :

- say : I want 2 components. Then your dimensionality reduction preserves 60 % + 30 % (general case: $\frac{\lambda_1 + \lambda_2}{\lambda_1 + \lambda_2 + \lambda_3} = \frac{\sum_1^p \lambda_i}{\sum_1^n \lambda_i}$, $(\lambda_i)$ in decreasing order) of the variance.


- say : I want to preserve 50 % of the variance. Then you can keep only 1 component (general case for $x$ % : $p_0 =  \mathrm{min}\{ p \leq n,   \frac{\sum_1^p \lambda_i}{\sum_1^n \lambda_i} \geq x/100 \} $
