##An informal, illustrated explanation of PCA in 2D and 3D


We are going to illustrate how a Principal Component Analysis work, on simple data in 2D and 3D.

PCA is often used to reduce the dimensionality of real-world data (for example, before applying clustering algorithms).


```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(1)


def generateEllipsis(mean, cov):
    data = np.random.multivariate_normal(mean, cov, 10000)
    lambd, axes = np.linalg.eig(cov)
    return data, axes, lambd


def plot2DEllipsis(data, axes):
    plt.axis("equal")
    plt.scatter(data[:,0], data[:, 1], color='red', s=1)    
    for i in range(len(axes)):
        plt.quiver(axes[0, i], axes[1, i], scale = 0.1, color='blue')       

        
def plot2DComponents(pca):
    components = pca.components_
    for comp in components:
       plt.quiver(comp[0], comp[1], scale=5, color='green')
```

Now we're going to generate pseudo random data (2D Gaussian). It'll be centered. For covariance matrix, we take 
$$\pmatrix{2 & -1\\-1 & 2}$$
If you want to try another ellipsis, note that the covariance matrix must be symmetrical, but also definite positive.


```python
mean = [0, 0]
cov1 = np.matrix([[2, -1], [-1, 2]])

ellipsis1, axes1, lambd1 = generateEllipsis(mean, cov1)
plotEllipsis(ellipsis1, axes1)

pca1 = PCA(n_components=2)
pca1.fit(ellipsis1)
plot2DComponents(pca1)

```

```python
print(pca1.explained_variance_ratio_) 
```

The first component is responsible for 75 % of the variance, and the second explains the 25 % remaining.

#General case

We need to define a 
For this we define an orthonormal matrix, P. In dimension 2 it is very easy: $$P = \frac{1}{a^2 + b^2}\pmatrix{a & -b \\ b &a}$$ 

Then $P^{-1} = P^{T}$, and with $D = \pmatrix{\lambda_1 & 0 \\ 0 & \lambda_2}$, we take as covariance matrix $PDP^{-1}$

```python
P = np.matrix([[1, -4], [4, 1]])
P = P/np.linalg.norm(P)
D = np.diag([1, 1])
cov2 = P*D*np.transpose(P)
print(np.transpose(P))
```

```python
ellipsis2, axes2, lambd2 = generateEllipsis(mean, cov2)

plot2DEllipsis(ellipsis2, axes2)

pca2 = PCA(n_components=2)
pca2.fit(ellipsis2)
plot2DComponents(pca2)
```

```python
print(pca2.explained_variance_ratio_)
```

The variance explained by the i-th component is proportional to $\lambda_i$ (where the eigenvalues are in descending order: $\lambda_1 \geq ... \geq \lambda_n$). Feel free to change the values of $D$ :)

In general there are two apporaches to PCA: 

- say "I want to keep $p$ components" (out of $n$ in the beggining, i.e. go from $\mathbb{R}^n$ to $\mathbb{R}^p$), and see how much of the variance is conserved 
- say "I want to keep xx% of the variance, without fixing the number of components.

It's not the best to see in 2D, so let's study it in 3D:
