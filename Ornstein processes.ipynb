# Ornstein processes : simulations and plotting

## Unidimensional

We first simulate one unidimensional Ornstein process, with chosen parameters :

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#1-dimension Ornstein Ulhenbeck process :
def simuOrnsteinOne(l, lambd, mu, sigma, dt, X0) :
    #length l, dX = lambd * (mu - X)dt + sigma * dB * sqrt(dt)
    X = np.zeros(l)
    X[0] = X0
    for t in range(1,l):
        X[t] = X[t-1] + lambd*(X[t-1]-mu)*dt + sigma*np.random.randn(1)*np.sqrt(dt)
    return np.asarray(X)
```

It gives us an array with the values of the process.
Now, we want to simulate n processes :

```python
#simulate n One-dimensional Ornstein processes of size l:
def simulationsOne(n, l):
    simus = []
    params = []
    for i in range(0,n):
        lambd = -np.random.uniform(0,1)
        mu = np.random.uniform(-1,1)
        sigma = np.random.uniform(-1,1)
        X0 = np.random.uniform(-1,1)
        X = simuOrnsteinOne(l, lambd, mu, sigma, 1, X0)
        simus.append(X)
        params.append([lambd, mu, sigma, X0])
    return np.asarray(simus), np.asarray(params)
```

It returns two arguments :
- the first one is a list, where each element is a 1-D process
- the second one is a list, where each element is the parameters of the process

Let's plot the results of the simulations :

```python
def plotSimulationsOne(L):
    fig= plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    for i in range(0,len(L)):      
        ax.plot(L[i])
        
S = simulationsOne(10,100)
plotSimulationsOne(S[0])
```

## Multidimensional

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#multi-dimensional Ornstein processes
def simuOrnsteinMulti(l, lambd, mu, sigma, dt, X0):
    #size l, parameters lambd and sigma are same-size squarred matrixes
    #mu and X0 vectors, dt scalar
    d = len(lambd) #dimension
    X=[np.asarray(X0)]
    for t in range(1,l):
        X.append(X[t-1] + np.dot(lambd,np.asarray(X[t-1]-mu))*dt + np.dot(sigma,np.random.randn(d)).tolist())
        
    params = [lambd, mu, sigma, X0]
    
    return np.asarray(X), params
    #a process is a matrix with l lines where each lines is a point in dim d
```

It return two arguments again :
- a list, where an element is a list of coordinates (a point of the process)
- a second list where an element is the parameters of one of the dimesion

We now simulate n processes :

```python
def simulationsMulti(n, l, d):
    #n samples of size l and dimension d
    simus = []
    params = []
    for i in range(0,n):
        lambd_diag = -np.diag(np.random.rand(d))
        P_plus = np.random.rand(d,d)
        P_minus = -np.random.rand(d,d)
        P = (P_plus + P_minus)
        lambd = np.dot(np.dot(P,lambd_diag), np.linalg.inv(P))
        
        sigma_plus = np.random.rand(d,d)
        sigma_minus = -np.random.rand(d,d)
        sigma = 10*(sigma_plus+sigma_minus)/2
        
        mu_plus = np.random.rand(d)
        mu_minus = -np.random.rand(d)
        mu = (mu_plus + mu_minus)
        
        X0_plus = np.random.rand(d)
        X0_minus = -np.random.rand(d)
        X0 = (X0_plus + X0_minus)
        
        X = simuOrnsteinMulti(l,lambd,mu,sigma,1,X0)
        simus.append(X[0])
        params.append(X[1])
        #simus : list of individual multi-dim processes
    return np.asarray(simus), params
```

- L is a list of multidim process
- params is the list of the list of params for the processes

In the cases d=1, 2 or 3, we can plot the results of the simulations :

```python
#S = simulationsMulti(n, l, d)[0]
def plotSimulationsMulti(S):
    n = len(S)
    d = len(S[0][0])
    if d >= 4:
        raise Exception("Too high dimension")
    elif d==1:
        plotSimulationsOne(S)
    elif d==2:
        S = np.asarray(S)
        fig= plt.figure()
        ax = fig.add_subplot(1, 1, 1)
        for i in range(0,n):      
            ax.plot(S[i][:,0], S[i][:,1])
    elif d==3:
        S = np.asarray(S)
        fig = plt.figure()
        ax = fig.gca(projection='3d')
        for i in range(0,n):    
            ax.plot(S[i][:,0],S[i][:,1],S[i][:,2])

#1D :
S = simulationsMulti(5, 100, 1)
plotSimulationsMulti(S[0])

#2D:
S = simulationsMulti(5, 100, 2)
plotSimulationsMulti(S[0])

#3D:
S = simulationsMulti(5, 100, 3)
plotSimulationsMulti(S[0])
```
