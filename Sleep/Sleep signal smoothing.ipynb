```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pylab import *
%matplotlib inline
```

```python
def get_signal(path_to_csv_file):
    sig = pd.read_csv(path_to_csv_file, squeeze=True).tolist()
    return sig
```

```python
def smooth(x,window_len=11,window='hanning'):
        if x.ndim != 1:
                raise ValueError, "smooth only accepts 1 dimension arrays."
        if x.size < window_len:
                raise ValueError, "Input vector needs to be bigger than window size."
        if window_len<3:
                return x
        if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
                raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
        s=np.r_[2*x[0]-x[window_len-1::-1],x,2*x[-1]-x[-1:-window_len:-1]]
        if window == 'flat': #moving average
                w=np.ones(window_len,'d')
        else:  
                w=eval('np.'+window+'(window_len)')
        y=np.convolve(w/w.sum(),s,mode='same')
        return y[window_len:-window_len+1]
```

```python
signal = get_signal("subject13_FPz_A1.csv")
len(signal)
```

There are two parameters to tune : the type of window we will use, and the length of the window.

```python
freq = 200 #512 points per second
span = 3 #3s window size

fig = plt.figure(figsize=(18, 12))

start = 3000600
sample = signal[start:(start + freq*span)] 
plt.plot(sample)

window_len = 15
smoothedFlat = smooth(np.asarray(sample), window_len=window_len, window='flat')
plt.plot(smoothedFlat, color ='green', linewidth = 2)

smoothedBartlett = smooth(np.asarray(sample), window_len=window_len, window='bartlett')
plt.plot(smoothedBartlett, color ='red', linewidth = 2)

smoothedHanning = smooth(np.asarray(sample), window_len=window_len, window='hanning')
plt.plot(smoothedHanning, color ='purple', linewidth = 2)
```

We try different smoothing kernels (see http://fr.wikipedia.org/wiki/Fen%C3%AAtrage#Fen.C3.AAtres_courantes)

Flat window does not perform well here compared to other windows : it tends to flatten the signal (fill gaps and cut peaks)
Bartlett window has a better visual result : it preserves gaps and peaks

```python
def getMinMaxList(sample):
    minList = np.asmatrix([(i, sample[i]) for i in range(1, len(sample) - 1) if  sample[i] <= sample[i-1] and sample[i] < sample[i+1]])
    maxList = np.asmatrix([(i, sample[i]) for i in range(1, len(sample) - 1) if  sample[i] > sample[i-1] and sample[i] >= sample[i+1]])
    return minList, maxList
```

```python
getMinMaxList(sample)
```

```python
def plotMinMax(sample, smoothed):
    fig = plt.figure(figsize=(18, 12))
    plt.plot(sample)
    plt.plot(smoothed, color ='orange', linewidth = 2, zorder = 1)

    minList, maxList = getMinMaxList(smoothed)
    plt.scatter(minList[:, 0], minList[:, 1], color = "green", zorder = 2)
    plt.scatter(maxList[:, 0], maxList[:, 1], color = "red", zorder = 3)
```

```python
smoothed = smooth(np.asarray(sample), window_len=window_len, window='flat')

plotMinMax(sample,smoothed)
```
